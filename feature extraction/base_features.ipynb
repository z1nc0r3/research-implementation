url length

```python
def length(url):
  return len(url)
```

protocol

```python
def protocol(url):
  if url.startswith('http://'):
    return 1
  elif url.startswith('https://'):
    return 2
  else: return 0
```

digits

```python
def digits(url):
  return len([char for char in url if char.isdigit()])
```

special chars

```python
def specials(url):
  return sum(1 for char in url if not char.isalnum())
```

dots and hyphes

```python
def dots(url):
  return url.count('.')
```

```python
def hyphens(url):
  return url.count('-')
```

slashes

```python
def single_slashes(url):
  return url.count('/')
```

```python
def double_slashes(url):
  return url.count('//')
```

@ sign

```python
def at_signs(url):
  return url.count('@')
```

% sign

```python
def percentages(url):
  return url.count('%')
```

**Run**

```python
df['commas'] = ''
df['tlides'] = ''
df['qms'] = ''
df['ands'] = ''
df['eqs'] = ''
df['underscores'] = ''
df['percentages'] = ''
df['www'] = ''
df.head()
```

```python
def extract():
  for i in df.index:
    url = df.iloc[i]['url']

    df.loc[i, 'commas'] = url.count(',')
    df.loc[i, 'tlides'] = url.count('~')
    df.loc[i, 'qms'] = url.count('?')
    df.loc[i, 'ands'] = url.count('&')
    df.loc[i, 'eqs'] = url.count('=')
    df.loc[i, 'underscores'] = url.count('_')
    df.loc[i, 'percentages'] = url.count('%')
    df.loc[i, 'www'] = url.count('www')

extract()
```
