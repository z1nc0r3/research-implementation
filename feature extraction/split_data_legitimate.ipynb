```python
def is_alive(url):
    try:
        res = requests.get(url, timeout=3)
        res.raise_for_status()
        return True
    except Exception as e:
        print(e)
        return False
```

```python
result = pd.DataFrame(columns=["url"])
```

```python
start = 0
```

```python
for i in range(start, len(legi)):
  url = legi.iloc[i]["url"]
  print(start, url)

  if is_alive(url):
    result.loc[start] = url


  start += 1
```
