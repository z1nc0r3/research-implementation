```python
from urllib.parse import urlparse, parse_qs
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Research content/data/2_legit.csv')
df.head()
```

```python
def extract_query_details(url):
    parsed_url = urlparse(url)

    query_params = parse_qs(parsed_url.query)

    return len(query_params)
```

```python
def is_subdirectory(url):
    parsed_url = urlparse(url)

    return 1 if bool(parsed_url.path) else 0
```

```python
df['is_subdir'] = ''
df.head()
```

```python
for i in df.index:
  url = df.iloc[i]['url']

  if url[-1] == '/':
    url = url[:len(url)-1]
  # print(extract_query_details(df.iloc[i]['url']))

  df.loc[i, 'is_subdir'] = is_subdirectory(url)
  print(i, "\r")
```
