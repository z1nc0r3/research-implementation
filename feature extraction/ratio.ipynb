```python
df = pd.read_csv('/content/drive/MyDrive/Research content/data/5_legit.csv')
df.head()
```

```python
def ratio(url):
  digits = sum(c.isdigit() for c in url)
  letters = sum(c.isalpha() for c in url)

  if digits == 0: return 0

  ratio = letters/digits
  return round(ratio, 2)
```

```python
df['domain_ratio'] = ''
df['ratio'] = ''
df.head()
```

```python
for i in df.index:
    url = df.iloc[i]['url']
    ext = tldextract.extract(url)
    domain = ext.fqdn

    df.loc[i, 'domain_ratio'] = ratio(domain)
    df.loc[i, 'ratio'] = ratio(url)
```
