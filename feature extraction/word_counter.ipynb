```python
import textstat
import pandas as pd
from zxcvbn import zxcvbn
```

```python
df = pd.read_csv('/content/drive/MyDrive/Research content/data/online_phish_features.csv')
df.head()
```

```python
def findTokens(results):
    for i in results['sequence']:
        if (i['pattern'] == 'dictionary') or (i['pattern'] == 'dictionary_reversed'):
            print(i['token'])

    return len(results['sequence'])
```

```python
df.insert(1, 'words', '')
```

```python
df.head()
```

```python
for i in range(len(df)):
    text = df['url'][i]
    count = findTokens(zxcvbn(text))
    df['words'][i] = count

    print(i, "\n")
```
