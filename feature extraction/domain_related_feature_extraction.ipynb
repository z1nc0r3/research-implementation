```python
df['domain_length'] = ''
df['domain_tld'] = ''
df['domain_digits'] = ''
df['domain_dots'] = ''
df['domain_hyphens'] = ''
df['domain_special'] = ''
```

```python
def domain_features():
  for i in range (df.size-1):
    url = df.iloc[i]['url']

    ext = tldextract.extract(url)
    domain = ext.fqdn
    suffix = ext.suffix

    print(i, ext.fqdn)

    df.loc[i, 'domain_length'] = len(domain)
    df.loc[i, 'domain_tld'] = suffix
    df.loc[i, 'domain_digits'] = len([char for char in domain if char.isdigit()])
    df.loc[i, 'domain_dots'] = domain.count('.')
    df.loc[i, 'domain_hyphens'] = domain.count('-')
    df.loc[i, 'domain_special'] = sum(1 for char in domain if not char.isalnum())

domain_features()
```

```python
df.to_csv('2_legit.csv', index=False)
```
