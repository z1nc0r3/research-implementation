```python
import pandas as pd
from zxcvbn import zxcvbn
```

```python
phish = pd.read_csv("/content/drive/MyDrive/Research content/data/online_phish_features.csv")
legit = pd.read_csv("/content/drive/MyDrive/Research content/data/online_legit.csv")
```

```python
phish.head()
```

```python
legit.head()
```

```python
def findTokens(results):
    return len(results['sequence'])
```

```python
legit['words'] = ''
phish['words'] = ''
```

```python
for i in range(len(phish)):
    text = phish['url'][i]
    count = findTokens(zxcvbn(text))
    phish['words'][i] = count

    print(i)
```
